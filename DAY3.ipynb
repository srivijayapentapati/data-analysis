##  geometric graph plot
##### ggplot also called as grammer of graph
-- grid plot

```python
from matplotlib import style
style.use("ggplot")
days=[1,2,3,4,5]
temperature=[36,35,37,40.6,38.5]
plt.xlabel("days",fontsize=15)
plt.ylabel("tempertaure",fontsize=10)
plt.plot(days,temperature,color="m",marker="*",markersize=8)
```

```python
vij_temp=[40.5,42.6,40.9,39.5,42.6]
plt.plot(days,vij_temp,'m*--',linewidth=3,label="vij temp")
plt.plot(days,temperature,color="g",marker="o",linewidth="4",label="gnt temp")
plt.title('gnt-tem vs vij-temp')
plt.xlabel("days")
plt.ylabel("temp")
plt.legend(loc=4)
plt.grid(color="r",linestyle="-",linewidth=10)
plt.show()
```

```python
semister=['sem1','sem2','sem3']
cgpa=[8.5,9.5,9.0]
plt.plot(semister,cgpa,color="r",linewidth=4,label="cgpa")
plt.title=("cgpa of a student")
plt.xlabel=("semisters")
plt.ylabel("cgpa")
plt.show()
```

```python
x=[9,3,5,4,8]
y=[10,6,4,5,4]
z=[7,8,9,7,4]
a=np.array([8,4,9,6,2])
plt.stackplot(x,y,z,a)
plt.show
```

```python
f=[1,2,3,4,5]
s={'y1':[5,6,7,8,9],'y2':[1,6,5,4,7],'y3':[2,4,6,7,1]}
plt.stackplot(f,s.values())
plt.show()
```

```python
plt.stackplot(f,s.values(),labels=s.keys())
plt.legend(loc='upper left')
plt.title("stack plot")
plt.show()
```
